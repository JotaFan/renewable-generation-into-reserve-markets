```python
def get_best_rows_on_lab(
    lab_results, validation_target, group_by, nn=False, max_archs=None, groupby_name="Arquitetura",do_the_mean=False,add_epoch=True,
):

    group_by_one = group_by
    if do_the_mean:
        group_by_one="name"
    res = lab_results.loc[
        lab_results.groupby(group_by_one)[validation_target].idxmax()
    ]
    #print(res)

    if do_the_mean:
        res = res[
            [group_by,"RMSE","SAE","AllocF","AllocD","GPD","GPD F","GPD D","GPD norm","GPD Positivo",  'GPD norm2']].groupby(group_by).mean()
        mean_series=res.index
        res.reset_index(inplace=True)
        res[group_by]=mean_series
        add_epoch=False


    
    cols_to_show =         [
            group_by,
            "RMSE",
            "SAE",
            "AllocF",
            "AllocD",
            "GPD",
            "GPD F",
            "GPD D",
            "GPD norm",
            "GPD Positivo",  'GPD norm2',
        ]
    if add_epoch:
        cols_to_show+=["epoch"]
    res = res[cols_to_show]
    if nn is not False:
        res["n"] = nn
    res = res.sort_values(
            by=validation_target, ascending=False
        )  # Example sorting, adjust as needed

    if max_archs is not None:
        # Assuming 'Arquitetura' is the column by which we determine the best architectures
        # Sort by 'Arquitetura' or any other metric you define as "best"
        res = res.head(max_archs)

    return res.rename(columns={group_by: groupby_name})
```

```python
def plot_the_results(best_rows, path_fig=None):

    
    
    cols_to_group = [['RMSE', 'SAE'], ['AllocF', 'AllocD'], ['GPD F', 'GPD D'], ['GPD', 'GPD norm2']]
    n_rows = len(cols_to_group)
    fig, axes = plt.subplots(nrows=n_rows, ncols=2, figsize=(20, 5*n_rows))
    
    for r in range(n_rows):
        ax_0 = axes[r, 0]
        ax_1 = axes[r, 1]
    
        col_to_show = cols_to_group[r]
        
        rows_to_show = best_rows[col_to_show[0]]
        if col_to_show[1] in benchmark_lines["Alocação a Descer"]:
            ax_1.axhline(y=benchmark_lines["Alocação a Descer"][col_to_show[1]], color='orange', linestyle='--',)
        if col_to_show[1] in benchmark_lines["Alocação a Subir"]:
            ax_1.axhline(y=benchmark_lines["Alocação a Subir"][col_to_show[1]], color='b', linestyle='--',)
        if col_to_show[0] in benchmark_lines["Alocação a Descer"]:
            ax_0.axhline(y=benchmark_lines["Alocação a Descer"][col_to_show[0]], color='orange', linestyle='--',)
        if col_to_show[0] in benchmark_lines["Alocação a Subir"]:
            ax_0.axhline(y=benchmark_lines["Alocação a Subir"][col_to_show[0]], color='b', linestyle='--',)
        
        
        
        subir_rows = rows_to_show.loc["Alocação a Subir"]
        descer_rows = rows_to_show.loc["Alocação a Descer"]
        rows_to_show = pd.DataFrame()
        rows_to_show["Alocação a Subir"]=subir_rows
        rows_to_show["Alocação a Descer"]=descer_rows
        
        rows_to_show.plot.bar(ax=ax_0)
    
        rows_to_show = best_rows[col_to_show[1]]
        subir_rows = rows_to_show.loc["Alocação a Subir"]
        descer_rows = rows_to_show.loc["Alocação a Descer"]
        rows_to_show = pd.DataFrame()
        rows_to_show["Alocação a Subir"]=subir_rows
        rows_to_show["Alocação a Descer"]=descer_rows
    
        rows_to_show.plot.bar(ax=ax_1)
    
        
        #if "GPD" in col_to_show:
         #   ax_0.set_ylim([-10,100])
          #  ax_1.set_ylim([-10,100])
    
    
        
        ax_0.set_title(col_to_show[0])
        ax_1.set_title(col_to_show[1])
        ax_1.set_xlabel(None)
        ax_0.set_xlabel(None)
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space at the top
    fig.suptitle(
        best_rows.index.names[-1],
        fontsize=16,
        y=0.98,
    )
    if path_fig is not None:
        fig.savefig(
            path_fig,
            bbox_inches="tight",
        )
```

```python
def get_filtered_lab_results(lab_results, exp_col,experiment_columns,count_min_val=3):
    non_exp_cols = [f for f in experiment_columns if f != exp_col]
    result_df = pd.DataFrame()
    for archs_value, group in lab_results.groupby(exp_col):
        # Select the non-experimental columns
        non_exp_series = group[non_exp_cols]
        
        # Create a Series of tuples with unique combinations
        combinations = pd.Series(non_exp_series.apply(tuple, axis=1)).drop_duplicates()
        combined_tuples = combinations.map(lambda x: (archs_value,) + x)
        
        # Convert the tuples back to a dataframe
        temp_df = pd.DataFrame(combined_tuples.tolist(), columns=[exp_col] + non_exp_cols)
        
        # Append the temporary dataframe to the result
        result_df = pd.concat([result_df, temp_df])
    result_df.dropna(inplace=True)
    result_df.reset_index(drop=True, inplace=True)
    result_df.sort_values(exp_col, inplace=True)
    counts = result_df.groupby(non_exp_cols)[exp_col].count().reset_index()
    counts =counts.sort_values(exp_col, ascending=False)
    print(counts[exp_col].unique())
    counts =counts[counts[exp_col]>count_min_val].reset_index(drop=True)
    filtered_lab_results=lab_results.merge(counts[non_exp_cols], on=non_exp_cols, how='inner')
    return filtered_lab_results
```

```python
# Function to extract the simple loss value
def extract_simple_loss(loss_value):
    # Convert to lowercase and remove non-alphanumeric characters
    if loss_value == "nan":
        return "nan"
    if loss_value != loss_value:
        return loss_value

    cleaned = ''.join(char for char in loss_value.lower() if char.isalnum())
    
    # Check if the cleaned string starts with any of the allowed values
    for value in allowed_loss_values:
        if cleaned.endswith(value):
            return value
    
    # If no match found, return None
    return None

# Apply the function to extract the simple loss value
lab_results_up['simple_loss'] = lab_results_up['loss'].apply(extract_simple_loss)
lab_results_down['simple_loss'] = lab_results_down['loss'].apply(extract_simple_loss)
```
